```python
from bs4 import BeautifulSoup
import requests
```

```python
url ="https://www.census2011.co.in/facts/highstateliteracy.html"
```

```python
page = requests.get(url)
```

```python
soup = BeautifulSoup(page.text, 'html')
print(soup)
```

```python
title = soup.find('title')
print(title)
```

```python
print(title.text)
```

```python
table = soup.find('table')
print(table)
```

```python
Headings = table.find_all('th')
print(Headings)
```

```python
Header = [i.text for i in Headings]
print(Header)
```

```python
import pandas as pd
df = pd.DataFrame(columns = Header)
df
```

```python
rows = table.find_all('tr')
print(rows)
```

```python
ind=0
for i in rows[1:6]:
  row_data = i.find_all('td')
  every_row_data = [i.text for i in row_data]
  print(every_row_data)
  df.loc[ind] = every_row_data
  ind = ind+1
```

```python
df
```

```python
df.to_csv("highstateliteracy.csv")
```

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url ="https://www.census2011.co.in/facts/highstateliteracy.html"
```

```python
p = requests.get(url)
```

```python
code = BeautifulSoup(p.text,'html')
```

```python
t = code.find('table')
print(t)
```

```python
data = t.find_all('td')
print(data)
```

```python
data_act = [i.text for i in data]
```

```python
print(data_act)
```

```python
newdf= pd.DataFrame(columns = ['Rank','Indian states','Literacy rate','men','women'])
```

```python
newdf
```

```python
data_act[0:4]
print(len(data_act))
```

```python
ind = 0
st = 0
stop = newdf.shape[1]
for i in data_act:
  if stop <= 40:
    newdf.loc[ind]= data_act[st:stop]
    st = st+stop
    stop = stop+stop
    ind = ind+1
```
